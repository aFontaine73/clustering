```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans 
from sklearn.cluster import MeanShift
from sklearn.datasets.samples_generator import make_blobs
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
```

```python
X = np.array([[1, 2],
             [1.5, 1.8],
             [5, 8],
             [8, 8],
             [1, 0.6],
             [9, 11]]
            )
```

```python
plt.scatter(X[: ,0], X[:, 1], s=150)
```

```python
clf = KMeans(n_clusters=2)
clf.fit(X)
```

```python
centroids = clf.cluster_centers_
centroids
```

```python
labels = clf.labels_
labels
```

```python
colours = ['g.', 'r.', 'c.', 'b.', 'k.', 'o.']
colours
```

```python
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], colours[labels[i]], markersize =25)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=150, linewidths=5)
```

```python
centres = [[1, 1], [5, 5]]
# X, _ = 
# samples generator will generate specific to machine learning sample data
# which means it'll come with labels
# normally don't have labeled data
# can be your y
# _ actual labels used in generation of dataset
X, _ = make_blobs(n_samples = 300, centers = centres, cluster_std = 1)
plt.scatter(X[:, 0], X[:, 1])
```

```python
ms = MeanShift()
ms.fit(X)
labels = ms.labels_
labels
```

```python
cluster_centres = ms.cluster_centers_
cluster_centres
```

```python
n_clusters_ = len(np.unique(labels))
n_clusters_
```

```python
colours = ['r','g','b','c','k','y','m'] * 10
colours
```

```python
for i in range(len(X)):
    plt.plot(X[i][0], X[i][1], color = colours[labels[i]], markersize =10)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(cluster_centres[:, 0], cluster_centres[:, 1], marker='x', s=150, linewidths=5, zorder=10)
```
